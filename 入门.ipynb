```python
#1 Âü∫Á°ÄÁªòÂõæ
#Á¨¨1Ê≠•ÔºöÂÆö‰πâxÂíåyÂùêÊ†áËΩ¥‰∏äÁöÑÁÇπ   xÂùêÊ†áËΩ¥‰∏äÁÇπÁöÑÊï∞ÂÄº
x=[1, 2, 3, 4]
#yÂùêÊ†áËΩ¥‰∏äÁÇπÁöÑÊï∞ÂÄº
y=[1, 4, 9, 16]
#Á¨¨2Ê≠•Ôºö‰ΩøÁî®plotÁªòÂà∂Á∫øÊù°Á¨¨1‰∏™ÂèÇÊï∞ÊòØxÁöÑÂùêÊ†áÂÄºÔºåÁ¨¨2‰∏™ÂèÇÊï∞ÊòØyÁöÑÂùêÊ†áÂÄº
plt.plot(x,y)
#Á¨¨3Ê≠•ÔºöÊòæÁ§∫ÂõæÂΩ¢
plt.show()


#2 ÂÆö‰πâÁªòÂõæÂ±ûÊÄß
'''
colorÔºöÁ∫øÊù°È¢úËâ≤ÔºåÂÄºrË°®Á§∫Á∫¢Ëâ≤ÔºàredÔºâ
markerÔºöÁÇπÁöÑÂΩ¢Áä∂ÔºåÂÄºoË°®Á§∫ÁÇπ‰∏∫ÂúÜÂúàÊ†áËÆ∞Ôºàcircle markerÔºâ
linestyleÔºöÁ∫øÊù°ÁöÑÂΩ¢Áä∂ÔºåÂÄºdashedË°®Á§∫Áî®ËôöÁ∫øËøûÊé•ÂêÑÁÇπ
'''
plt.plot(x, y, color='r',marker='o',linestyle='dashed')
#plt.plot(x, y, 'ro')
'''
axisÔºöÂùêÊ†áËΩ¥ËåÉÂõ¥
ËØ≠Ê≥ï‰∏∫axis[xmin, xmax, ymin, ymax]Ôºå
‰πüÂ∞±ÊòØaxis[xËΩ¥ÊúÄÂ∞èÂÄº, xËΩ¥ÊúÄÂ§ßÂÄº, yËΩ¥ÊúÄÂ∞èÂÄº, yËΩ¥ÊúÄÂ§ßÂÄº]
'''
plt.axis([0, 6, 0, 20])
plt.show()


#3 ÁªôÂõæÁâáÊèêÈòøÁî≤Ê≥®ÈáäÂíåÊ†áÈ¢òÁ≠â

#Á¨¨1Ê≠•ÔºöÂÆö‰πâxÂíåyÂùêÊ†áËΩ¥‰∏äÁöÑÁÇπ  xÂùêÊ†áËΩ¥‰∏äÁÇπÁöÑÊï∞ÂÄº
x=[1, 2, 3, 4]
#yÂùêÊ†áËΩ¥‰∏äÁÇπÁöÑÊï∞ÂÄº
y=[1, 4, 9, 16]

#Á¨¨2Ê≠•Ôºö‰ΩøÁî®plotÁªòÂà∂Á∫øÊù° Á¨¨1‰∏™ÂèÇÊï∞ÊòØxÁöÑÂùêÊ†áÂÄºÔºåÁ¨¨2‰∏™ÂèÇÊï∞ÊòØyÁöÑÂùêÊ†áÂÄº
plt.plot(x,y)
#Ê∑ªÂä†ÊñáÊú¨ #xËΩ¥ÊñáÊú¨
plt.xlabel('xÂùêÊ†áËΩ¥')
#yËΩ¥ÊñáÊú¨
plt.ylabel('yÂùêÊ†áËΩ¥')
#Ê†áÈ¢ò
plt.title('Ê†áÈ¢ò')
#Ê∑ªÂä†Ê≥®Èáä ÂèÇÊï∞ÂêçxyÔºöÁÆ≠Â§¥Ê≥®Èáä‰∏≠ÁÆ≠Â§¥ÊâÄÂú®‰ΩçÁΩÆÔºåÂèÇÊï∞ÂêçxytextÔºöÊ≥®ÈáäÊñáÊú¨ÊâÄÂú®‰ΩçÁΩÆÔºå
#arrowpropsÂú®xyÂíåxytext‰πãÈó¥ÁªòÂà∂ÁÆ≠Â§¥, shrinkË°®Á§∫Ê≥®ÈáäÁÇπ‰∏éÊ≥®ÈáäÊñáÊú¨‰πãÈó¥ÁöÑÂõæÊ†áË∑ùÁ¶ª

plt.annotate('ÊàëÊòØÊ≥®Èáä', xy=(2,5), xytext=(2, 10),
            arrowprops=dict(facecolor='black', shrink=0.01),
            )

#Á¨¨3Ê≠•ÔºöÊòæÁ§∫ÂõæÂΩ¢
plt.show()
```
