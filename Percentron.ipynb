```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1=[0.2,0.4,1,1.5]
x2=[0.4,0.5,0.8,1.5]
x3=[1,1,1,1]
y=[-1,-1,1,1]
w1=0
w2=0
w3=0
yita=0.1
```

```python
def percention(w1,w2,w3,x1,x2,x3):
    if(w1*x1+w2*x2+w3*x3<=0):
        return -1
    else:
        return 1
```

```python
for i in range(1000):
    for j in range(4):
        if(percention(w1,w2,w3,x1[j],x2[j],x3[j])!=y[j]):
            if(percention(w1,w2,w3,x1[j],x2[j],x3[j])>0):
                w1 = w1 -  yita*x1[j]
                w2 = w2 -  yita*x2[j] 
                w3 = w3 -  yita*x3[j] 
            else:               
                w1 = w1 +  yita*x1[j]
                w2 = w2 +  yita*x2[j]
                w3 = w3 +  yita*x3[j] 
    print(w1,w2,w3)
print(w1,w2,w3)
```

```python
for i in range(4):
    print(percention(w1,w2,w3,x1[i],x2[i],x3[i]))
```
